```python
import sys
import sklearn
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Fetching the iris dataset
from sklearn import datasets

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
# Loading to pandas Dataframe
df = pd.DataFrame(features)
df.columns = iris.feature_names

print(df.head(20))
print(df.describe())
```

```python
print(df.shape)
print(target.shape)
print(target)
print(iris.target_names)
```

```python
# Displaying scatter plot matrix
from pandas.plotting import scatter_matrix

scatter_matrix(df, figsize = (10, 10))
plt.show()
```

```python
# Determining optimal number of clusters using Elbow method
from sklearn.cluster import KMeans

X = []  # Clusters
Y = []  # awcss

for i in range(1, 32):
    kmeans = KMeans(n_clusters = i);
    kmeans.fit(df)
    
    # Calculate average within cluster sum of squares i.e. inertia wrt every point
    awcss = kmeans.inertia_ / df.shape[0]
    
    X.append(i)
    Y.append(awcss)
```

```python
# Plotting X and Y

plt.plot(X, Y, 'bo-')
plt.xlim((1, 31))
plt.xlabel('No. of clusters')
plt.ylabel('Averge within-Clusters Sum of Squares')
plt.title('KMeans Clustering Elbow Method')

plt.show()
```

## Principal Component Analysis (PCA)

```python
# Reducing dataset to 2 features so that it can be visualized on a plot
from sklearn.decomposition import PCA
from sklearn import preprocessing

pca = PCA(n_components = 2)
c_f = pca.fit_transform(df)

print(c_f.shape)
print(c_f[:10])
```

```python
# Fit the model to compressed data
kmeans = KMeans(n_clusters = 3)
kmeans.fit(c_f)
```

```python
# Visualizing high dimensional clusters using principal components

# Setting size for mesh
h = 0.01

# Generating mesh grid
x_min, x_max = c_f[:, 0].min() - 1, c_f[:, 0].max() + 1
y_min, y_max = c_f[:, 1].min() - 1, c_f[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Label each point in mesh using last trained model(on PC)
# Obtaining points
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Generate colour plot
Z = Z.reshape(xx.shape)
plt.clf()
plt.figure(figsize = (12, 12))
plt.imshow(Z, interpolation = 'nearest', extent = (xx.min(), xx.max(), yy.min(), yy.max()), 
          cmap = plt.cm.tab20c, aspect = 'auto', origin = 'lower')

# Plotting principal components
for i, point in enumerate(c_f):
    if target[i] == 0:
        plt.plot(point[0], point[1], 'g.', markersize = 10)
    elif target[i] == 1:
        plt.plot(point[0], point[1], 'r.', markersize = 10)
    else:
        plt.plot(point[0], point[1], 'b.', markersize = 10)
        
# Plotting cluster centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'x', s = 250, linewidth = 4, color = 'w', zorder = 10)

# Setting plot
plt.title('KMeans Clustering on PCA reduced Iris Dataset')
plt.xlim((x_min, x_max))
plt.ylim((y_min, y_max))
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.xticks(())
plt.yticks(())

plt.show()
```

## Comparing Metrics
* Homogeneity - measures whether or not all of its clusters contain only data points which are members of a single class.
* Completeness - measures whether or not all members of a given class are elements of the same cluster
* V-measure - the harmonic mean between homogeneity and completeness

```python
from sklearn import metrics

# Model for original dataset
kmeans1 = KMeans(n_clusters = 3)
kmeans1.fit(df)

# Model for compressed dataset
kmeans2 = KMeans(n_clusters = 3)
kmeans2.fit(c_f)

# Metrics
print('\nOriginal data')
print('Homogeneity: {}'.format(metrics.homogeneity_score(target, kmeans1.labels_)))
print('Completeness: {}'.format(metrics.completeness_score(target, kmeans1.labels_)))
print('V-Measure: {}'.format(metrics.v_measure_score(target, kmeans1.labels_)))

print('\nPCA reduced data')
print('Homogeneity: {}'.format(metrics.homogeneity_score(target, kmeans2.labels_)))
print('Completeness: {}'.format(metrics.completeness_score(target, kmeans2.labels_)))
print('V-Measure: {}'.format(metrics.v_measure_score(target, kmeans2.labels_)))
```

```python
print('Original: {}'.format(kmeans1.labels_))
print('PCA Reduced: {}'.format(kmeans2.labels_))
print('Actual: {}'.format(target))
```
